```python
#NORTH AMERICAN
```

```python
import os
import nltk
from nltk.corpus import stopwords
import string
from wordcloud import WordCloud
import matplotlib.pyplot as plt

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
custom_stop_words = {'f.', 'would', 'little', 'zone', 'b.', 'j', 'et'}
stop_words.update(custom_stop_words)

directory_path = '/Users/ashlynnmorin/desktop/classes/PlantandPython/HISTORICAL_PAPERS/NorthAmerica'  # Replace with the actual directory path
all_texts = []

txt_files = [file for file in os.listdir(directory_path) if file.endswith('.txt')]

for file_name in txt_files:
    file_path = os.path.join(directory_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        all_texts.append(text)

combined_text = '\n'.join(all_texts)

words = nltk.word_tokenize(combined_text)
filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word not in string.punctuation]

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_words))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Combined Word Cloud for All Files')
plt.show()
```
